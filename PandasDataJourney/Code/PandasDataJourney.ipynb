### Creating DataFrames

```python
df = pd.DataFrame() # Empty DataFrame
```

```python
df = pd.DataFrame([[1, 2, 3], [4, 5, 6], [7, 8, 9]]) # Basic Matrix
```

```python
df = pd.DataFrame(
    [[1, 2, 3], [4, 5, 6], [7, 8, 9]],
    columns=["A", "B", "C"],
    index=["X", "Y", "Z"], copy=False
) # With Custom Columns and Index and Preventing Data Copy
```
