## Cleaning Data

### Cleaning Empty Cells

#### Remove Rows

```python
df = pd.read_csv('Data.csv')

df.dropna(inplace = True)

print(df.to_string())
```

#### Replace Empty Values

```python
df.fillna(130, inplace = True)
```

#### Replace Only For Specified Columns

```python
df.fillna({"Calories": 130}, inplace=True)
```

#### Replace Using Mean, Median, or Mode

```python
x = df["Calories"].mean()
y = df["Calories"].median()
z = df["Calories"].mode()[0]
# exemple replace with mean value
df.fillna({"Calories": x}, inplace=True)
```

### Cleaning Data of Wrong Format

#### Data of Wrong Format

##### Convert Into a Correct Format

```python
df['Date'] = pd.to_datetime(df['Date'], format='mixed')
```

#### Replacing Values

```python
df.loc[7, 'Duration'] = 45
```

```python
print(df)
```

### Removing Duplicates

```python
print(df.duplicated())
```

```python
df.drop_duplicates(inplace = True)
```

```python
print(df)
```

## Advanced Data Filtering and Querying Methods

```python
bios=pd.read_csv('bios.csv')
```

##### Conditional Filtering

```python
bios.loc[bios['height_cm'] > 220]
```

```python
bios.loc[
    (bios['height_cm'] > 215) & 
    (bios["born_country"] == "USA"), 
    ["name", "height_cm"]
]
```

##### String Operation Filtering

```python
bios[bios['name'].str.contains("Keith")]
```

##### SQL-Style Query Interface

```python
bios.query('born_country == "USA"')
```
